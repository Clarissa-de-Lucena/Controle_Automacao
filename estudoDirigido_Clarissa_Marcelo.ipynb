# Capitulo 4: Caracteristicas de Sistema de Controle com retroação

A retroação é um elemento essencial nos sistemas de controle, e o sinal de erro desempenha um papel central nesse mecanismo. O principal objetivo é reduzir esse sinal de erro para aproximar o desempenho real ao desejado. Além disso, a retroação auxilia na atenuação dos efeitos de variações indesejadas nos parâmetros do sistema, contribuindo para a manutenção de sua estabilidade e precisão. Outro ponto relevante é a supressão das perturbações na saída do sistema, o que resulta em maior consistência nos resultados. Vale destacar que o custo de implementação de um sistema de retroação está diretamente relacionado ao tipo de sensor utilizado, uma vez que este sensor é responsável por medir as variáveis necessárias para o controle adequado do sistema.

Essa reformulação se alinha ao capítulo , onde o papel da retroação no controle de sistemas dinâmicos, estabilidade, e a mitigação de distúrbios são amplamente abordados.

### Sistemas de malha aberta e fechada

Um sistema de controle pode ser classificado como de malha aberta ou de malha fechada, dependendo da existência ou não de retroação.

No caso de um sistema de malha aberta (ou sistema direto), o controle ocorre sem a consideração do feedback ou da resposta real do sistema. O sinal de controle é gerado unicamente com base nas variáveis de entrada e na função de transferência do sistema. Um exemplo típico de sistema de malha aberta é o controle de uma lâmpada através de um interruptor. Ao acionar o interruptor, a lâmpada se acende, independentemente das condições de iluminação do ambiente.

Por outro lado, em um sistema de malha fechada (ou sistema com retroação), o controle é baseado no feedback. Um sinal de erro, que reflete a diferença entre a saída desejada e a saída real, é gerado e utilizado para ajustar o sinal de controle, resultando em um desempenho mais preciso e estável. Um exemplo clássico de sistema de malha fechada é o controle de temperatura em um forno. Sensores medem a temperatura real e comparam com o valor desejado. Com base na diferença, o sistema ajusta a quantidade de energia fornecida para manter a temperatura dentro do valor estipulado.

Assim, enquanto um sistema de malha aberta opera sem considerar o feedback da saída, um sistema de malha fechada utiliza a retroação para ajustar continuamente o controle, visando alcançar o desempenho desejado.

### Sensibilidade de sistemas de controle a variação de parâmetros

Um sistema de controle enfrenta desafios como o envelhecimento do processo, a falta de conhecimento sobre os valores exatos dos parâmetros e outros fatores naturais que podem afetar o procedimento de controle. Em sistemas de malha aberta, esses erros resultam em modificações e imprecisões na saída, tornando a sensibilidade do sistema às variações de parâmetros uma questão fundamental.

Uma maneira útil de lidar com isso é aumentar a magnitude da função de transferência da malha, representada por G(s)H(s). Assim, a principal vantagem de um sistema com retroação é a redução das variações nos parâmetros do processo (G(s)).

A sensibilidade do sistema é definida como a relação entre a mudança na função de transferência e a mudança na função de transferência do processo (ou parâmetro) para uma pequena mudança incremental. Em outras palavras, mede a resposta do sistema a variações nos parâmetros.

Portanto, ao adicionar um loop de controle, é possível reduzir o efeito das variações nos parâmetros de um sistema de controle, o que se torna crucial nos sistemas de controle com retroação.

O conceito de sensibilidade também pode ser encontrado em amplificadores operacionais e está relacionado à capacidade de resposta do amplificador a mudanças nos componentes e parâmetros do circuito.






### Controle da resposta transitória de sistemas

A resposta transitória de um sistema é a resposta do sistema ao longo do tempo. Em um sistema de controle, a resposta transitória é influenciada pelos parâmetros e configuração da malha de controle.

Em um sistema de malha fechada, é possível ajustar os parâmetros da malha de retroação para obter a resposta desejada. Isso permite um maior controle sobre a resposta transitória, buscando um tempo de resposta mais rápido, menor oscilação ou outras características desejadas.

No caso de um sistema de malha aberta, onde não há retroação, é possível modificar a resposta transitória através da inserção de um controlador em cascata adequado. Isso significa adicionar um componente adicional ao sistema para influenciar sua resposta.

Se a resposta transitória de um sistema, seja em malha aberta ou fechada, for muito lenta e não atender aos requisitos desejados, uma opção é escolher outro componente do sistema, como um motor, com uma constante de tempo diferente. Isso implica substituir o elemento que está causando a lentidão por um com uma resposta mais rápida, se possível, a fim de melhorar a resposta transitória do sistema.

Em resumo, a resposta transitória de um sistema pode ser ajustada através dos parâmetros da malha de controle em um sistema de malha fechada ou pela escolha adequada de componentes em um sistema de malha aberta, buscando alcançar uma resposta mais rápida e adequada aos requisitos do sistema.






### Sinais de perturbação em um sistema de controle com retroação

Primeiramente, é importante dizer que: um sinal de perturbação é um sinal de entrada indesejável que afeta o sinal de saída do sistema. É um sinal que acontece em diversos sistemas de controle, produzindo uma saída inexata.

Uma das principais razões de se inserir a retroação é a capacidade de diminuir os efeitos de perturbações e de ruídos que ocorrem dentro da malha de retroação.
É evidente que o esforço para reduzir os efeitos de variações de parâmetros ou de perturbações são proporcionais, e elas se reduzem simultaneamente.

### Erro de estado estacionário

O sistema de controle com retroação é importante, pois oferece ao engenheiro o recurso de ajustar a resposta transitória. Deve-se realizar uma comparação do erro final de estado estacionário de um sistema a malha aberta com o de um sistema a malha fechada.
O erro de estado estacionário é o erro depois que a resposta transitória desapareceu, deixando apenas a resposta contínua.


Para calcular o erro de estado estacionário, se utiliza o teorema do valor final, ou seja:

$$ \lim_{t \rightarrow \infty} e(t) =\lim_{s \rightarrow 0} sE(s)  $$


Um sistema de controle em malha fechada reduz o erro estacionário causado por alterações nos parâmetros ou erros de calibração. Exemplos incluem controle térmico, regulação de tensão e controle de nível de água.

### Custos da retroação

O custo da retroação se manifesta primeiro no aumento do número de componentes e na complexidade do sistema. O sensor é um elemento chave para a retroação. O sensor geralmente é o componente mais caro em um sistema de controle.

O segundo custo da retroação é a perda de ganho. Corriqueiramente, existe ganho a malha aberta com sobra, e se deseja trocar isto por um aumento no controle da resposta do sistema.

Outro custo da retroação é a introdução da possibilidade de instabilidade. Embora o sistema a malha aberta seja estável, o sistema a malha fechada pode não ser sempre estável.

Resposta para malha fechada

```python
import numpy as np
import matplotlib.pyplot as plt
from control import tf, feedback, step_response

def simulate_step_response(Ka, t):
    G = tf([1], [1, 20, 0])              # G(s) = 1 / (s^2 + 20s)
    F = tf([5000], [1, 1000])            # F(s) = 5000 / (s + 1000)
    Gc = Ka * F                          # Gc(s) = Ka * F(s)
    Gol = Gc * G                         # Sistema em malha aberta
    T = feedback(Gol, 1)                 # Malha fechada unitária
    t, y = step_response(T, t)
    return t, y

# Tempo de simulação
t = np.linspace(0, 2, 201)

# Respostas para Ka = 10 e Ka = 80
t1, y1 = simulate_step_response(10, t)
t2, y2 = simulate_step_response(80, t)

# Plotagem em subplots, como na figura
fig, axs = plt.subplots(2, 1, figsize=(8, 6), sharex=True)

axs[0].plot(t1, y1, 'b')
axs[0].set_ylabel('y(t)')
axs[0].grid(True)
axs[0].set_title('Figura 4.35(b) - Resposta ao Degrau para $K_a = 10$')
axs[0].set_xlim([0, 2])
axs[0].set_ylim([0, 1])

axs[1].plot(t2, y2, 'r')
axs[1].set_xlabel('Tempo (s)')
axs[1].set_ylabel('y(t)')
axs[1].grid(True)
axs[1].set_title('Figura 4.35(b) - Resposta ao Degrau para $K_a = 80$')
axs[1].set_xlim([0, 2])
axs[1].set_ylim([0, 1.2])

plt.tight_layout()
plt.show()
```

Resposta a uma perturbação em degrau

```python
import numpy as np
import matplotlib.pyplot as plt
from control import tf, feedback, step_response

def simulate_disturbance_response(Ka, t):
    G = tf([1], [1, 20, 0])              # Planta: 1 / (s^2 + 20s)
    F = tf([5000], [1, 1000])            # Filtro rápido
    Gc = Ka * F                          # Controlador proporcional
    Gol = Gc * G                         # Sistema em malha aberta

    # Malha fechada considerando entrada como perturbação (negativa)
    T = feedback(G, Gc)                  # Planta com realimentação do controlador
    T_neg = -T                           # Perturbação negativa
    t, y = step_response(T_neg, t)       # Resposta ao degrau da perturbação
    return t, y

# Simulação
t = np.linspace(0, 2, 201)
t, y = simulate_disturbance_response(Ka=80, t=t)

# Plot
plt.figure(figsize=(8, 4))
plt.plot(t, y * 1e3, 'b')  # Multiplica por 1e3 para exibir como na figura (x10⁻³)
plt.title('Fig. 4.35 - Resposta à Perturbação em Degrau para $K_a = 80$')
plt.xlabel('Tempo (s)')
plt.ylabel('y(t) [$\\times 10^{-3}$]')
plt.grid(True)
plt.xlim([0, 2])
plt.ylim([-3, 0])
plt.tight_layout()
plt.show()
```

# Capitulo 5: O desempenho de sistemas de controle com reatroação

Uma característica vantajosa dos sistemas de controle com retroação é a capacidade de ajustar o desempenho tanto em regime transitório quanto em regime permanente. Ao analisar e projetar um sistema de controle, é necessário definir uma medida para o desempenho desejado. Com base nesse objetivo, os valores dos parâmetros do sistema podem ser ajustados para alcançar a resposta desejada.

Os sistemas de controle são dinâmicos por natureza e, portanto, seu desempenho é especificado em termos da resposta transitória e da resposta em regime estacionário. A resposta transitória é a parte da resposta que diminui ao longo do tempo, enquanto a resposta em regime estacionário é a parte da resposta que ocorre muito tempo após a aplicação do sinal de entrada.

Ao projetar sistemas de controle, as especificações incluem vários índices que medem a resposta temporal para um determinado sinal de entrada, bem como a precisão desejada em regime permanente. Esses índices ajudam a determinar o quão rápido o sistema responde às mudanças e quão próximo ele está do valor desejado em condições estacionárias.

Em resumo, ao projetar sistemas de controle com retroação, é importante considerar tanto a resposta transitória quanto a resposta em regime estacionário, levando em conta as especificações de desempenho desejadas para o sistema.

Um sistema é considerado um sistema de controle ótimo quando os parâmetros do sistema são ajustados de modo que o índice alcance um valor extremo, comumente um valor mínimo. Um índice de desempenho,para ser útil, deve ser um número sempre positivo ou nulo. Então, o melhor sistema é definido como o sistema que minimiza este índice.

```python
import numpy as np
import matplotlib.pyplot as plt
from control import tf, feedback, step_response

def simulate_step_response(Ka, t):
    G = tf([1], [1, 20, 0])
    Gc = tf([Ka * 5], [1])
    Gol = Gc * G
    T = feedback(Gol, 1)
    t, y = step_response(T, t)
    return t, y

t = np.linspace(0, 1, 100)

t1, y1 = simulate_step_response(30, t)
t2, y2 = simulate_step_response(60, t)

plt.figure()
plt.plot(t1, y1, label='$K_a = 30$')
plt.plot(t2, y2, label='$K_a = 60$')
plt.xlabel('Tempo (s)')
plt.ylabel('y(t)')
plt.title('Resposta ao Degrau para Diferentes Valores de $K_a$')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

Resposta do Sistema a uma perturbação em degrau unitário, D(s) = 1/s

```python
import numpy as np
import matplotlib.pyplot as plt
from control import tf, feedback, step_response

def simulate_perturbation_response(Ka, t, label):
    C = tf([Ka * 5], [1])
    G = tf([1], [1, 20, 0])
    T = feedback(G, C)

    num_closed = T.num[0][0]
    den_closed = T.den[0][0]

    num_closed = [-x for x in num_closed]
    system_neg = tf(num_closed, den_closed)

    t_out, y_out = step_response(system_neg, t)
    plt.plot(t_out, y_out, label=label)

t = np.arange(0, 1.01, 0.01)

plt.figure()
simulate_perturbation_response(30, t, r'$K_a=30$')
simulate_perturbation_response(60, t, r'$K_a=60$')

plt.grid(True)
plt.xlabel('Tempo (s)')
plt.ylabel('y(t)')
plt.title('Resposta para K = 30 e 60')
plt.legend()
plt.tight_layout()
plt.show()
```

# Capitulo 6: A estabilidade de sistemas lineares com retroação

Ao abordar o projeto e a análise de sistemas de controle com retroação, a estabilidade é um aspecto fundamental. Um exemplo claro é o de caças modernos, que são projetados para serem instáveis em malha aberta. Sem o controle ativo com retroação, que auxilia o piloto, esses aviões não seriam capazes de voar. Os engenheiros implementam o controle ativo para estabilizar o comportamento inerentemente instável da aeronave, permitindo que outros aspectos, como o desempenho em regime transitório, possam ser ajustados.

A retroação possibilita a estabilização de processos instáveis por meio da escolha criteriosa de parâmetros, regulando a performance durante mudanças de estado.

A estabilidade absoluta refere-se à condição em que um sistema com retroação é classificado como estável ou instável. Se o sistema apresentar uma resposta controlada e não crescer indefinidamente, ele é considerado estável.

Um sistema é dito estável quando, ao ser submetido a uma perturbação, sua resposta permanece limitada em magnitude. Além disso, um sistema é considerado dinâmico quando sua resposta é limitada para uma entrada também limitada.

Essa reformulação está em conformidade com os conceitos do capítulo onde a estabilidade, tanto absoluta quanto relativa, é tratada com ênfase, junto à importância da retroação na estabilização de sistemas naturalmente instáveis.

### O critério de estatibilidade de Routh-Hurwitz

A estabilidade de um sistema é determinada pelo critério de Routh-Hurwitz, que analisa o polinômio característico do sistema. Um sistema é considerado instável se todos os coeficientes desse polinômio forem positivos.

O critério de Routh-Hurwitz é um método necessário e suficiente para avaliar a estabilidade de sistemas lineares. Inicialmente, foi desenvolvido utilizando determinantes, mas posteriormente foi apresentada uma forma mais conveniente usando uma tabela organizada. O critério é baseado na ordem dos coeficientes do polinômio e utiliza um algoritmo que calcula esses coeficientes, baseando-se em determinantes de matrizes ou na forma da equação diferencial.

De acordo com a metodologia de Routh-Hurwitz, o número de raízes da equação polinomial com parte real positiva é igual ao número de mudanças de sinal na primeira coluna da tabela de Routh. Portanto, para que um sistema seja estável, não deve haver mudanças de sinal na primeira coluna. Esse requisito é simultaneamente necessário e suficiente para a estabilidade do sistema.

Através de rearranjos matemáticos, são considerados quatro casos diferentes que podem ocorrer nas equações expressas no tópico:

Nenhum elemento da primeira coluna é igual a zero;
Existe um zero na primeira coluna, mas outros elementos da linha que contém esse zero são diferentes de zero;
Existe um zero na primeira coluna e os outros elementos da linha também são iguais a zero;
O mesmo caso que o anterior, mas com raízes repetidas no eixo jw.
Esses casos são analisados para determinar a estabilidade do sistema de acordo com o critério de Routh-Hurwitz.

### A estabilidade relativa de sistemas de controle com retroação

Entre as várias definições da estabilidade relativa aqui encontramos duas bastante importantes:

-  A estabilidade relativa de um sistema pode ser definida como uma propriedade que é medida pela parte real relativa de cada raiz ou par de raízes;

-  A estabilidade relativa de um sistema também pode ser definida em termos dos coeficientes de amortecimento.

E nessa relação entre as duas raízes complexas encontradas, temos uma certa velocidade na resposta obtida e no tempo da ultrapassagem, que sobrepõe o tempo de assentamento. Por isso é importante uma análise criteriosa dos parâmetros do polinômio que rege esses sistemas de estabilidade relativa.

```python
import control as ct
import numpy as np
import matplotlib.pyplot as plt

Ka = 100
K1 = 0.05

sys1 = ct.TransferFunction([5000], [1, 1000])
sys2 = ct.TransferFunction([1], [1, 20, 0])
sysc = ct.TransferFunction([K1, 1], [1])

syso = ct.series(Ka * sys1, sys2)

sys = ct.feedback(syso, sysc)
sys = ct.minreal(sys)

t = np.arange(0, 0.5, 0.001)
t, y = ct.step_response(sys, t)

plt.figure(figsize=(12, 6))
plt.plot(t, y,color='red')
plt.grid(True)
plt.xlabel('Time (s)')
plt.ylabel('y(t)')
plt.title('Resposta do sistema com retroação de velocidade - Ka = 100 e K1 = 0.05')
plt.xlim(0, 0.5)
plt.ylim(0, 1)
plt.show()
```

# Capitulo 7: O método do lugar das raízes

O desempenho dinâmico de um sistema de controle em malha fechada é descrito pela função de transferência em malha fechada, que é representada por uma razão de polinômios. Essa função, denotada por T(s), relaciona a saída do sistema (Y(s)) com o sinal de referência (R(s)).

A equação característica do sistema, representada por 1 + KG(s) = 0, é fundamental para determinar os modos de resposta do sistema. Nessa equação, K é um parâmetro variável e G(s) é a função de transferência do processo.

As raízes da equação característica, localizadas no plano s, determinam os modos de resposta do sistema. Para que o sistema seja estável, as raízes devem satisfazer a condição em que |KG(s)| = 1. Isso significa que o produto do ganho K e da função de transferência G(s) deve ter uma magnitude igual a 1.

Além disso, a fase do produto KG(s) deve ser igual a -180° (ou 180° mais um múltiplo inteiro de 360°) para garantir a estabilidade do sistema. Essa condição é expressa na forma polar, onde o ângulo de KG(s) deve ser um múltiplo de 180°.

Em resumo, para que um sistema de controle em malha fechada seja estável, é necessário que o produto KG(s) tenha magnitude igual a 1 e uma fase de -180° (ou 180° mais um múltiplo inteiro de 360°). Essas condições garantem a localização adequada das raízes características do sistema no plano s.






**Sistema de leitura de acionador de disco (Exemplo 7.11)**

```python
import numpy as np
import matplotlib.pyplot as plt
import control

G = control.TransferFunction((1, 1.5), (1, 11, 10, 0))
gains = np.linspace(0.0, 100.0, num=1000)

# Calcula o root locus sem plotar
rlist, klist = control.rlocus(G, gains=gains, plot=False)

# Cria figura personalizada
fig, ax = plt.subplots(figsize=(12, 8))

# Plota os ramos do lugar das raízes
for i in range(rlist.shape[1]):
    ax.plot(rlist[:, i].real, rlist[:, i].imag, 'b')

# Plota polos e zeros
poles = np.roots(G.den[0][0])
zeros = np.roots(G.num[0][0])

ax.plot(poles.real, poles.imag, 'rx', markersize=10, label='Polos')
ax.plot(zeros.real, zeros.imag, 'go', markersize=10, label='Zeros')

# Ajustes de grid
ax.set_title('Root Locus')
ax.set_xlabel('Parte Real')
ax.set_ylabel('Parte Imaginária')

ax.legend()
plt.xlim(-10, 0)
plt.ylim(-8, 8)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct

K = [50, 100, 200, 500]

numg = [1]
deng = [1, 15, 50, 0]
sysg = ct.tf(numg, deng)

t = np.arange(0, 5, 0.1)

responses = []

for k in K:
    sys = ct.feedback(k * sysg)
    t, y = ct.step_response(sys, t)
    responses.append(y)

plt.figure(figsize=(10, 6))
for i, k in enumerate(K):
    plt.plot(t, responses[i], label=f'K = {k}')

plt.xlabel('Time (s)')
plt.ylabel('y(t)')
plt.title('Resposta transitória para o controlador de ganho simples')
plt.grid(True)
plt.legend()
plt.show()
```

Também introduz o projeto de sistemas de controle usando o método de variáveis de estado, que oferece uma perspectiva mais abrangente para a modelagem de sistemas complexos.

O capítulo se encerra com um exemplo prático de um controlador proporcional e derivativo (PD) que inclui filtragem prévia, aplicado a um sistema de leitura de acionador de disco. Nesse exemplo, são analisados os desafios e soluções específicas para obter um controle eficiente e estável, reforçando a aplicabilidade dos conceitos discutidos ao longo do texto.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

Kp = 2.0  # Ganho proporcional
Kd = 1.0  # Ganho derivativo
tau_filter = 0.1  # Constante de tempo para a filtragem prévia

mass = 1.0  # Massa do acionador
damping = 0.5  # Damping
spring_constant = 10.0  # Constante da mola

def system_dynamics(y, t, u):
    position, velocity = y
    acceleration = (u - damping * velocity - spring_constant * position) / mass
    return [velocity, acceleration]

def pd_control(setpoint, position, velocity, previous_error, dt):
    error = setpoint - position
    derivative = (error - previous_error) / dt
    filtered_derivative = (1/tau_filter) * (derivative + previous_error)

    u = Kp * error + Kd * filtered_derivative
    return u, error

time = np.linspace(0, 10, 1000)  # Tempo de simulação
setpoint = 1.0  # Ponto de ajuste (posição desejada)
y0 = [0.0, 0.0]  # Condições iniciais: [posição, velocidade]

positions = []
velocities = []
errors = []
previous_error = 0.0
dt = time[1] - time[0]

for t in time:
    u, previous_error = pd_control(setpoint, y0[0], y0[1], previous_error, dt)

    y0 = odeint(system_dynamics, y0, [0, dt], args=(u,))[-1]

    positions.append(y0[0])
    velocities.append(y0[1])
    errors.append(setpoint - y0[0])

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(time, positions, label='Posição', color='b')
plt.axhline(setpoint, color='r', linestyle='--', label='Ponto de ajuste')
plt.title('Posição do Acionador de Disco')
plt.xlabel('Tempo (s)')
plt.ylabel('Posição')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(time, velocities, label='Velocidade', color='g')
plt.title('Velocidade do Acionador de Disco')
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(time, errors, label='Erro', color='orange')
plt.axhline(0, color='k', linestyle='--')
plt.title('Erro de Controle')
plt.xlabel('Tempo (s)')
plt.ylabel('Erro')
plt.legend()

plt.tight_layout()
plt.show()
```

O código simula um sistema de controle que utiliza um controlador proporcional e derivativo (PD) com filtragem prévia, aplicado a um acionador de disco. Inicialmente, são definidos os parâmetros essenciais, como os ganhos do controlador e características do sistema, incluindo massa, amortecimento e constante da mola.

A dinâmica do acionador é modelada por uma função que calcula a aceleração com base na posição e na velocidade atuais, além do sinal de controle. Em seguida, uma função para o controlador PD é implementada. Essa função mede o erro entre a posição desejada e a atual, calcula a taxa de variação desse erro (derivada) e aplica uma filtragem prévia para suavizar a resposta, evitando oscilações indesejadas.

Com as funções definidas, o código executa uma simulação ao longo de um intervalo de tempo, atualizando continuamente o sinal de controle. A cada iteração, o controlador ajusta a saída com base no erro calculado, e as dinâmicas do sistema são integradas para obter as novas posições e velocidades do acionador.

Durante essa simulação, os valores de posição, velocidade e erro são armazenados em listas. Por fim, o código gera gráficos que mostram a evolução da posição do acionador ao longo do tempo, sua velocidade e o erro de controle. Isso permite visualizar a eficácia do controlador em alcançar e manter a posição desejada, destacando como o sistema responde ao sinal de controle ao longo do tempo.

A resposta de um sistema de segunda ordem com dados amostrados a uma entrada degrau.

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct

T = 1

num = [0.3678, 0.3678 * 0.7189]
den = [1, -1.3678, 0.3678]

G = ct.tf(num, den, dt=T)

T_sys = ct.feedback(G, 1)

t_discreto, y_discreto = ct.step_response(T_sys, T=15)

G_cont = ct.tf([1], [1, 1, 0])
T_cont = ct.feedback(G_cont, 1)
t_cont = np.linspace(0, 15, 1000)
t_cont, y_cont = ct.step_response(T_cont, t_cont)

plt.figure(figsize=(12, 6))

markerline, stemlines, baseline = plt.stem(
    t_discreto,
    y_discreto,
    linefmt='b-',
    markerfmt='bo',
    basefmt=' ',
    label='Sistema Discreto (T = 1)'
)

plt.setp(stemlines, linewidth=2)
plt.setp(markerline, markersize=8)

plt.plot(t_cont, y_cont, 'r-', label='Sistema Contínuo')

plt.title('Resposta ao Degrau - Sistemas Discreto vs. Contínuo')
plt.xlabel('Tempo (segundos)')
plt.ylabel('Amplitude da Saída')
plt.legend()
plt.grid(True)
plt.axis([0, 15, 0, 1.5])
plt.show()

print("Análise do Sistema Discreto:")
print(f"- Sobressinal (Overshoot): {100*(np.max(y_discreto) - 1):.1f}%")
print(f"- Tempo de Acomodação (2%): {t_discreto[np.where(np.abs(y_discreto - 1) > 0.02)[0][-1]]:.2f} s")

print("\nAnálise do Sistema Contínuo:")
print(f"- Sobressinal (Overshoot): {100*(np.max(y_cont) - 1):.1f}%")
print(f"- Tempo de Acomodação (2%): {t_cont[np.where(np.abs(y_cont - 1) > 0.02)[0][-1]]:.2f} s")
```
